```python
import pandas as pd
import matplotlib.pyplot as plt
```

<b> Broad Analysis Skills</b>

```python
df = pd.read_excel('Global Superstore.xls')
df.columns
```

```python
mask_brazil = df['Country'] =='Brazil'
df_brazil = df[mask_brazil]
df_brazil
```

```python
df_brazil['City'].value_counts()
```

```python
df_sales = df_brazil.groupby('City')[['Sales']].sum().reset_index().sort_values(by='Sales' ,ascending=False).head(10)
df_sales
```

```python
df_cost = df_brazil.groupby('City')[['Shipping Cost']].sum().reset_index().sort_values(by='Shipping Cost' ,ascending=False).head(10)
df_cost
```

```python
df_profit = df_brazil.groupby('City')[['Profit']].sum().reset_index().sort_values(by='Profit' ,ascending=False).head(10)
df_profit
```
